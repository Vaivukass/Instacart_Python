## 3. If Statements & loc Function

```python
# Defining a function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function

df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Checking price ranges

df['price_range'].value_counts(dropna = False)
```

```python
# The most expensive product

df['prices'].max()
```

```python
# Trying to use loc function to achieve the same as if-else statement

df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

```python
# Loc function on the entire df

df_ords_prods_mer.loc[df_ords_prods_mer['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_ords_prods_mer.loc[(df_ords_prods_mer['prices'] <= 15) & (df_ords_prods_mer['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

## 4. For-loops

```python
# Check counts for days of the week

df_ords_prods_mer['order_day_of_week'].value_counts(dropna = False)
```

```python
# Create a for-loop and start with an empty list for the result

result = []

for value in df_ords_prods_mer["order_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

### Step 4: Create a new column 'busiest_period_of_day' with 3 values: 'Most orders', 'Average orders', 'Fewest orders'

```python
# Check counts for order_hour_of_day

df_ords_prods_mer['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Create a for-loop for orders

result_orders = []

for value in df_ords_prods_mer['order_hour_of_day']:
  if value == 10 or value == 11 or value == 14 or value == 15 or value == 13:
    result_orders.append('Most orders')
  elif value == 3 or value == 4 or value == 2 or value == 5 or value == 1:
    result_orders.append('Fewest orders')
  else:
    result_orders.append('Average orders')
```

```python
result_orders
```

```python
# Create a column to store the above result

df_ords_prods_mer['busiest_period_of_day'] = result_orders
```

```python
# Get counts

df_ords_prods_mer['busiest_period_of_day'].value_counts(dropna = False)
```
