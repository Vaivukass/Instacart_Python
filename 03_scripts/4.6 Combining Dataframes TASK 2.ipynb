```python
# Importing products_checked.csv

df_prods = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'products_checked.csv'), index_col = False)
```

## 3. Exploring Data

```python
# Removing eval_set column

df_ords_prods_comb = df_ords_prods_comb.drop(columns = ['eval_set'])
```

```python
# Dropping the additional index column "Unnamed:0"

df_prods = df_prods.drop(columns = ['Unnamed: 0'])
```

## 4. Merging Dataframes

```python
# Merge df_ords_prods_comb and df_prods

df_merged = df_prods.merge(df_ords_prods_comb, on = 'product_id', indicator = 'ind')
```

```python
df_merged.head()
```

```python
# Check counts of ind

df_merged['ind'].value_counts()
```

```python
# Doing an outer merge to test if we have a full merge

df_merged_outer = df_prods.merge(df_ords_prods_comb, on = 'product_id', how = 'outer', indicator = 'ind')
```

```python
df_merged_outer.head()
```

```python
# Check counts of ind

df_merged_outer['ind'].value_counts()
```

```python
# Remove _merge column

df_merged = df_merged.drop(columns = ['_merge'])
```

## 5. Exporting Dataframe

```python
# Exporting orders_products_merged.pkl

df_merged.to_pickle(os.path.join(path, '02 Data','Prepared Data', 'orders_products_merged.pkl'))
```
